# Preprocessing

### Load Dataset

```python
tips = sns.load_dataset("tips")  
tips.tail()
```

### Add Columns

- 분석 시 필요한 컬럼 추가 : 주문 총액 대비 팁 비율 (`tip(%)`)

```python
tips["tip(%)"] = (tips["tip"] / tips["total_bill"] * 100).round(2)
print(tips.shape)
tips.head()
```

### 3. Identify & drop Outlier

```python
tips.loc[tips['tip(%)'] > 45]
```

```python
tips = tips.drop(tips.index[172])
```

```python
tips.loc[tips['tip(%)'] > 45]
print(tips.shape)
```

### Rearrange columns and save to CSV

```python
tips_2 = tips[["day", "time", "sex", "size", "smoker", "total_bill", "tip", "tip(%)"]]
print(tips_2.shape)
tips_2.head()
```

```python
tips_2.to_csv('tips_2.csv')
```

# Hypothesis & Analysis

### 가설 (1)

주말에 남성이 지불 할 때, 팁을 더 후하게 줄 것이다.

- 분석 방법:
    - boxplot을 통해 과 시간대 별 팁 지불 비율을 비교한다

```python
sns.boxplot(x="day", y="tip(%)", hue="sex", 
            data=tips_2)
plt.show()
```

결과 : 비교 결과 큰 차이가 없는 것으로 나타났다. 가설 재설정

### 가설 (2)
흡연자가 금액을 지불 할 경우, 팁의 비율과 상관관계가 있을 것이다.

방법 : boxplot을 통해 흡연 여부와 팁의 관계를 파악한다.

```python
sns.boxplot(x="smoker", y="tip(%)", data=tips_2)
plt.show()
```

결과 : 명확한 차이는 없어보이며, 흡연을 할 경우 지불하는 총액 대비 팁을 계산하는 정도의 변동 폭이 크다. 

### 가설 (3)

2인 단위 손님의 팁 지불 금액이 다른 그룹의 손님들보다 더 높을 것이다.

- 분석 방법 : 
    - swarmplot을 통해 각 그룹사이즈 별 팁 금액을 비교해본다.

```python
sns.swarmplot(x='size', y='tip(%)', data=tips_2, hue='time')
```

결과 : 총 주문 금액 대비 25%이상 팁을 지불한 경우 또한 다른 그룹보다 많으므로,가설에 부합하는 것을 확인 할 수 있다.
