```python
import random, string

def get_name():
    names = ["Jooen", "Ken", "Jody", "Peter", "Emily", "Henry", "Ben"]
    return random.choice(names)

def get_age(start=20, end=50):
    return random.randint(start, end)

def make_data (rows=10):
    datas = []
    for _ in range(rows):
        data = {"Age": get_age(), "Name": get_name()}
        datas.append(data)
    return datas
```

```python
employee_df = pd.DataFrame(columns=["Employee_No", "Name", "Age"])

for idx in range(1, 6+1):
    name = get_name()
    
    while name in list(employee_df["Name"]):
        name = get_name()
        
    data = {"Name":name, "Employee_No":idx, "Age":get_age()}
    employee_df.loc[len(employee_df)] = data
    
    
employee_df
```

```python
salary_df = pd.DataFrame(columns=["ID", "Salary"])

for idx in range(1, 20):
    salary = random.randint(1, 20)*1000
    data = {"Salary":salary, "ID":random.randint(1, 6+1)}
    salary_df.loc[len(salary_df)] = data
    
salary_df
```

```python
result_df = salary_df.merge(employee_df, left_on="ID", right_on="Employee_No")
result_df
```
