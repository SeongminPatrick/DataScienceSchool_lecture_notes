```python
from sklearn.datasets import load_digits
digits = load_digits()

X = digits.images[0].reshape(1, 64)
for num in range(1, 1797):
    d = digits.images[num].reshape(1, 64)
    X = np.r_[X, d]
    
    
X

```

```python
X = pd.DataFrame(X)
X.head()
```

```python
sim_X = np.dot(X, X.T)
print(sim_X.shape)
print(sim_X)
```

```python
plt.imshow(sim_X, cmap="Greys");
plt.colorbar()
plt.show()
```

```python
X_heat = sns.heatmap(sim_X, cmap="Greys")
X_heat
```

```python
corr = np.corrcoef(X, X.T)
corr
```
